# Identificando objetos con MobileNetV2

```python
# ==========================================
# 1. INSTALACIÓN E IMPORTACIÓN
# ==========================================
# Instalamos gymnasium (Numpy ya viene preinstalado en Colab)
!pip install gymnasium

import gymnasium as gym
import numpy as np
import random

# Se inicia el entorno usando FrozenLake-v1.
env = gym.make('FrozenLake-v1', is_slippery=True)

# ==========================================
# 2. HIPERPARÁMETROS DEL AGENTE
# ==========================================
# Se definen los hiperparametros de aprendizaje
num_states = env.observation_space.n
num_actions = env.action_space.n

# Inicializamos la tabla Q con ceros
q_table = np.zeros((num_states, num_actions))

# Parámetros de aprendizaje
total_episodes = 2000
learning_rate = 0.9          # Alpha
discount_factor = 0.95       # Gamma

# Parámetros de exploración (Epsilon-Greedy)
epsilon = 1.0
max_epsilon = 1.0
min_epsilon = 0.01
decay_rate = 0.005

# ==========================================
# 3. ENTRENAMIENTO CON Q-LEARNING
# ==========================================
print("Entrenando al agente... Por favor espera.")

for episode in range(total_episodes):
    state, info = env.reset()
    done = False

    while not done:
        # Selección de acción: Exploración vs. Explotación
        exp_exp_tradeoff = random.uniform(0, 1)

        if exp_exp_tradeoff > epsilon:
            action = np.argmax(q_table[state, :]) # Explotación
        else:
            action = env.action_space.sample()    # Exploración

        # Ejecutar acción
        next_state, reward, terminated, truncated, info = env.step(action)
        done = terminated or truncated

        # Ecuación de Bellman para actualizar la Tabla Q
        q_table[state, action] = q_table[state, action] + learning_rate * (
            reward + discount_factor * np.max(q_table[next_state, :]) - q_table[state, action]
        )

        state = next_state

    # Reducir epsilon para explorar cada vez menos
    epsilon = min_epsilon + (max_epsilon - min_epsilon) * np.exp(-decay_rate * episode)

print("¡Entrenamiento completado!\n")

# ==========================================
# 4. EVALUACIÓN DEL RENDIMIENTO
# ==========================================

env_eval = gym.make('FrozenLake-v1', is_slippery=True, render_mode="ansi")
eval_episodes = 100
successful_episodes = 0

for episode in range(eval_episodes):
    state, info = env_eval.reset()
    done = False

    while not done:
        action = np.argmax(q_table[state, :]) # Solo mejores acciones
        state, reward, terminated, truncated, info = env_eval.step(action)
        done = terminated or truncated

        if terminated and reward == 1.0:
            successful_episodes += 1

env_eval.close()

# Mostrar resultados
success_rate = (successful_episodes / eval_episodes) * 100
print(f"--- RESULTADOS DE LA EVALUACIÓN ---")
print(f"Episodios exitosos: {successful_episodes}/{eval_episodes}")
print(f"Tasa de éxito: {success_rate}%")
```
