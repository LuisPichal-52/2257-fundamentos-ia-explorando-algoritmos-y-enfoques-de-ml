```python
import os
from google.colab import userdata
os.environ["GEMINI_API_KEY"] = userdata.get("Alura_Curso")
from google import genai

def resumir_lista_emails(lista_emails):
    client = genai.Client()
    resúmenes = []

    print("Procesando y resumiendo correos con IA...\n")

    for n, email in enumerate(lista_emails, 1):
        prompt = f"""
        Analiza el siguiente correo electrónico y haz un resumen de su contenido principal
        en una sola línea, de forma clara y directa.

        Correo:
        {email}
        """

        try:
            response = client.models.generate_content(
                model='gemini-2.5-flash',
                contents=prompt
            )

            resumen_limpio = response.text.strip().replace('\n', ' ')
            resúmenes.append(resumen_limpio)

            print(f"Email {n}: {resumen_limpio}")

        except Exception as e:
            print(f"Error al procesar el email {n}: {e}")
            resúmenes.append("Error al generar el resumen.")

    return resúmenes

# Ejecutamos la función
lista_de_resumenes = resumir_lista_emails(ejemplos_emails)
```
